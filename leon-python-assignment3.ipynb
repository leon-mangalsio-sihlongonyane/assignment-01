(a) Program for the Runge-kutta four-order (RK4) method to approximate the population at $t = 50$ using
step size $h = 0.01$. We evaluate for $f_{5000}$.

<html><font color = 'red', size = 4><br>

```python
k=20   # initiating constants.
h=0.01
r=0.4
f = lambda y : r*y*(1-(y/K)) - y**2/(1 +y**2)  #defining the density function 
t,y = 0,2.44  #initiating t,y to able to use in the FOR LOOP.
T=[0]    #initiating point on the list
Y=[2.44]  # initiating list with intial population = 2.44, then append from here onward
for i in range(8000): #FOR LOOP repeat process below for 8000 times (picked as part of (b))
    k1 = h*f(y)      #RK4 algorithm: k1,k2,k3,k4, t and y.
    k2 = h*f(y+0.5*k1)
    k3 = h*f(y+0.5*k2)
    k4 = h*f(y+k3)
    t+= h  
    y += k1/6 + k2/3 + k3/3 + k4/6   
    T.append(t)  #appending t each time to list T to obtain list of values of T in 8000 run.
    Y.append(y)      #appending y each time to list Y to obtain list of values of Y in 8000 run.
print(Y[5000])  #gives approximate value of the population at t = 50, i.e 5000th value of Y (population).
```
